# Gaussian Process

```python
import torch
import gpytorch
import numpy as np

# ============================================================
# ---- Base Exact GP model -------------------
# ============================================================

class ExactGPModel(gpytorch.models.ExactGP):
    def __init__(self, train_x, train_y, likelihood, kernel='RBF', ard_dims=None):
        super().__init__(train_x, train_y, likelihood)
        self.mean_module = gpytorch.means.ConstantMean()

        # Choose kernel type
        if kernel == 'RBF':
            base_kernel = gpytorch.kernels.RBFKernel(ard_num_dims=ard_dims)
        elif kernel == 'Matern':
            base_kernel = gpytorch.kernels.MaternKernel(nu=2.5, ard_num_dims=ard_dims)
        elif kernel == 'RQ':
            base_kernel = gpytorch.kernels.RationalQuadraticKernel(ard_num_dims=ard_dims)
        else:
            raise ValueError(f"Unsupported kernel type: {kernel}")

        self.covar_module = gpytorch.kernels.ScaleKernel(base_kernel)

    def forward(self, x):
        mean_x = self.mean_module(x)
        covar_x = self.covar_module(x)
        return gpytorch.distributions.MultivariateNormal(mean_x, covar_x)


# ============================================================
# ---- Individual GP Manager: handles data + normalization ----
# ============================================================

class GPManager:
    def __init__(self, kernel='RBF', lr=0.03, iters=1000):
        self.kernel = kernel
        self.lr = lr
        self.iters = iters
        self.trained = False
        self.X_train = []
        self.Y_train = []

        self.likelihood = None
        self.model = None


    # ----------------------------- #
    #        FIT / INITIAL TRAIN    #
    # ----------------------------- #
    def fit(self, X, Y):
        X = torch.tensor(X, dtype=torch.float32)
        Y = torch.tensor(Y, dtype=torch.float32)

        self.X_train = X.clone()
        self.Y_train = Y.clone()

        self.retrain()

    def retrain(self):
        # Normalize the input/output and then train the model
        self._compute_normalization()
        self._train_model()

    # ----------------------------- #
    #       ADD NEW DATA POINTS     #
    # ----------------------------- #
    def add_data(self, X_new, Y_new, retrain=True):
        """
        Add new data (single or batch) to GP training set.
        Optionally retrain the model.
        """
        X_new = torch.tensor(X_new, dtype=torch.float32)
        Y_new = torch.tensor(Y_new, dtype=torch.float32).flatten()  # ensure 1D

        # Also flatten stored Y to ensure shape consistency
        if self.Y_train.ndim > 1:
            self.Y_train = self.Y_train.flatten()

        # Append new data
        self.X_train = torch.cat([self.X_train, X_new], dim=0)
        self.Y_train = torch.cat([self.Y_train, Y_new], dim=0)

        if retrain:
            self.retrain()

    # ----------------------------- #
    #         INTERNAL UTILS        #
    # ----------------------------- #
    def _compute_normalization(self):
        self.X_mean, self.X_std = self.X_train.mean(0), self.X_train.std(0)
        self.Y_mean, self.Y_std = float(self.Y_train.mean(0)), float(self.Y_train.std(0))

        Xn = (self.X_train - self.X_mean) / self.X_std
        Yn = (self.Y_train - self.Y_mean) / self.Y_std

        self.Xn, self.Yn = Xn, Yn

    def dataset(self):
        # Extract training data (already stored inside the wrapper)
        X_train = self.X_train.numpy()
        Y_train = self.Y_train.numpy()
        return X_train, Y_train


    def _train_model(self):
        self.likelihood = gpytorch.likelihoods.GaussianLikelihood()
        self.likelihood.noise_covar.initialize(noise=1e-3)
        self.model = ExactGPModel(self.Xn, self.Yn, self.likelihood,
                                  kernel=self.kernel, ard_dims=self.X_train.shape[-1])

        # -------------------------------------------------
        # Train the model
        # -------------------------------------------------
        self.train_gp(self.model, self.likelihood, self.Xn, self.Yn)
        self.trained = True

    def train_gp(self, model, likelihood, x, y):
        model.train(); likelihood.train()
        opt = torch.optim.Adam(model.parameters(), lr=self.lr)
        mll = gpytorch.mlls.ExactMarginalLogLikelihood(likelihood, model)
        for _ in range(self.iters):
            opt.zero_grad()
            out = model(x)
            loss = -mll(out, y)
            loss.backward()
            opt.step()
        model.eval(); likelihood.eval()

    def predict(self, X):
        if not self.trained:
            raise RuntimeError("GP has not been trained yet.")

        # Normalize the input
        X = torch.as_tensor(X, dtype=torch.float32)
        Xn = (X - self.X_mean) / self.X_std
        with torch.no_grad(), gpytorch.settings.fast_pred_var():
            pred = self.likelihood(self.model(Xn))

            # Unonormalize the output
            mean = pred.mean.cpu().numpy() * self.Y_std + self.Y_mean
            var = pred.variance.cpu().numpy() * (self.Y_std ** 2)
        # Return the unnormalized output
        return mean, var
```

# Setup environment and collect initial data

```python
import gymnasium as gym
# import gym
import numpy as np
import torch
import matplotlib.pyplot as plt

# --------------------------------------------------------------
# 1.  Environment setup
# --------------------------------------------------------------
env = gym.make("MountainCarContinuous-v0", max_episode_steps=500)
n_state = env.observation_space.shape[0]    # [position, velocity]
n_action = env.action_space.shape[0]        # 1 continuous action
n_output = 2                                # Δpos, Δvel


# --------------------------------------------------------------
# 2.  Collect initial random data
# --------------------------------------------------------------
data = {'X': [], 'Y': []}

obs, _ = env.reset(seed=0)
for _ in range(500):  # collect 500 transitions
    a = env.action_space.sample()  # random continuous action in [-1,1]
    x_next, r, term, trunc, _ = env.step(a)
    data['X'].append(np.concatenate([obs, a]))   # [pos, vel, action]
    data['Y'].append(x_next - obs)               # Δstate
    obs = x_next
    if term or trunc:
        obs, _ = env.reset()

X = np.array(data['X'])[::3]
Y = np.array(data['Y'])[::3]
print(f"Collected {len(X)} transitions.")


# --------------------------------------------------------------
# 3.  Train GP dynamics models
# --------------------------------------------------------------
gps = [GPManager(kernel='RBF', iters=500) for _ in range(n_output)]

for d in range(n_output):
    gps[d].fit(X, Y[:, d])
    print(f"Trained GP for Δstate[{d}] with {len(X)} samples.")
```

## COST FUNCTION

```python
import numpy as np

# Goal position from the environment
goal_pos = 0.45

def stage_cost(x, u):
    x = np.asarray(x)
    u = np.asarray(u)

    pos = x[..., 0]
    vel = x[..., 1]

    # Penalize being left of the goal, large velocities, and large actions
    dist_to_goal = np.maximum(0.0, goal_pos - pos)
    cost_pos = dist_to_goal**2          # strongly encourage reaching the goal
    cost_vel = 0.1 * vel**2             # keep velocity moderate
    cost_u   = 0.001 * u**2             # small penalty on control effort

    return cost_pos + cost_vel + cost_u


def trajectory_cost(xs, us):

    xs = np.asarray(xs)
    us = np.asarray(us)
    assert xs.shape[0] == us.shape[0] + 1, "xs should have one more step than us"

    return float(np.sum(stage_cost(xs[:-1], us)))
```

## MPPI ACTION

```python
import numpy as np
import matplotlib.pyplot as plt

# Horizon and MPPI hyperparameters
H       = 80           # prediction horizon
K       = 1000          # number of sampled trajectories
LAMBDA  = 1.0          # temperature
SIGMA   = 0.6          # std dev of exploration noise
rng     = np.random.default_rng(0)

env = gym.make("MountainCarContinuous-v0", render_mode="rgb_array")

# Action bounds from environment
act_low  = env.action_space.low[0]
act_high = env.action_space.high[0]


def gp_step_batch(states, actions):

    states = np.asarray(states, dtype=np.float32)
    actions = np.asarray(actions, dtype=np.float32)

    # Build GP inputs [pos, vel, action]
    X = np.column_stack([states[:, 0], states[:, 1], actions])

    dpos_mean, dpos_var = gps[0].predict(X)
    dvel_mean, dvel_var = gps[1].predict(X)

    next_states = np.empty_like(states)

    next_states[:, 0] = states[:, 0] + dpos_mean
    next_states[:, 1] = states[:, 1] + dvel_mean

    # Clip to valid observation space
    low  = env.observation_space.low
    high = env.observation_space.high
    next_states = np.clip(next_states, low, high)

    return next_states


def mppi_action(x0, u_init=None):

    x0 = np.asarray(x0, dtype=np.float32)

    if u_init is None:
        u_init = np.zeros(H, dtype=np.float32)
    else:
        u_init = np.asarray(u_init, dtype=np.float32)
        assert u_init.shape[0] == H

    # Sample noise around current plan
    eps = rng.normal(0.0, SIGMA, size=(K, H)).astype(np.float32)
    U   = np.clip(u_init[None, :] + eps, act_low, act_high)

    # Rollout trajectories under GP model
    states = np.tile(x0, (K, 1))      # (K, 2)
    costs  = np.zeros(K, dtype=np.float32)

    for t in range(H):
        u_t = U[:, t]                 # (K,)
        # stage_cost is from the previous cell
        costs += stage_cost(states, u_t)
        states = gp_step_batch(states, u_t)

    # MPPI weighting
    J_min   = costs.min()
    weights = np.exp(-(costs - J_min) / LAMBDA)
    weights_sum = weights.sum() + 1e-8

    # Weighted update of the control sequence
    du     = (weights[:, None] * eps).sum(axis=0) / weights_sum
    u_new  = np.clip(u_init + du, act_low, act_high)
    u0     = float(u_new[0])

    return u0, u_new
```

### RUN SIMULATION

```python
from IPython.display import display, clear_output

def run_mppi_episode(seed=0, max_steps=200, render=True):
    obs, _ = env.reset(seed=seed)
    u_plan = np.zeros(H, dtype=np.float32)

    positions = []
    rewards   = []

    fig, ax = plt.subplots()
    img = ax.imshow(env.render()); ax.axis("off"); display(fig)

    done, t = False, 0

    #for t in range(max_steps):
    while not done and t < max_steps:
        positions.append(obs[0])

        # Get MPPI action
        u0, u_plan = mppi_action(obs, u_plan)

        # Apply to real env
        obs, r, terminated, truncated, _ = env.step(np.array([u0], dtype=np.float32))
        rewards.append(r)

        if render:
            img.set_data(env.render())
            ax.set_title(f"t={t} | x={obs[0]:+.3f} v={obs[1]:+.3f} a={u0:+.3f}")
            clear_output(wait=True); display(fig); plt.pause(0.001)
            t += 1
        # Shift plan (receding horizon)
        u_plan = np.roll(u_plan, -1)
        u_plan[-1] = 0.0

        if terminated or truncated:
            break

    return np.array(positions), np.array(rewards)




positions, rewards = run_mppi_episode(seed=0, max_steps=500, render=False)
print(f"Steps: {len(positions)}, Return: {rewards.sum():.2f}")

plt.figure(figsize=(6, 4))
plt.plot(positions, label="position")
plt.axhline(env.unwrapped.goal_position, linestyle="--", color="r", label="goal")
plt.xlabel("time step")
plt.ylabel("position")
plt.title("Exercise 1 – GP-based MPPI using stage_cost")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

# COST FUNCTION MEAN-VARIANCE AWARE

```python
import numpy as np

# Goal position from the environment
goal_pos = 0.45

def stage_cost(x, u, uncertainty, use_var):
    x = np.asarray(x)
    u = np.asarray(u)

    pos = x[..., 0]
    vel = x[..., 1]

    # Penalize being left of the goal, large velocities, and large actions
    dist_to_goal = np.maximum(0.0, goal_pos - pos)
    cost_pos = 100*dist_to_goal**2          # strongly encourage reaching the goal
    cost_vel = 0.1 * vel**2             # keep velocity moderate
    cost_u   = 0.001 * u**2             # small penalty on control effort

    cost_uncer = use_var*uncertainty

    return cost_pos + cost_vel + cost_u + cost_uncer
```

# MPPI THAT SEND MEAN AND VARIANCE

```python
import numpy as np
import matplotlib.pyplot as plt

# Horizon and MPPI hyperparameters
H       = 80           # prediction horizon
K       = 1000          # number of sampled trajectories
LAMBDA  = 1.0          # temperature
SIGMA   = 0.6          # std dev of exploration noise
rng     = np.random.default_rng(0)

env = gym.make("MountainCarContinuous-v0", render_mode="rgb_array")

# Action bounds from environment
act_low  = env.action_space.low[0]
act_high = env.action_space.high[0]


def gp_step_batch(states, actions):

    states = np.asarray(states, dtype=np.float32)
    actions = np.asarray(actions, dtype=np.float32)

    # Build GP inputs [pos, vel, action]
    X = np.column_stack([states[:, 0], states[:, 1], actions])

    dpos_mean, dpos_var = gps[0].predict(X)
    dvel_mean, dvel_var = gps[1].predict(X)

    uncertainity = dpos_var + dvel_var

    next_states = np.empty_like(states)

    next_states[:, 0] = states[:, 0] + dpos_mean
    next_states[:, 1] = states[:, 1] + dvel_mean

    # Clip to valid observation space
    low  = env.observation_space.low
    high = env.observation_space.high
    next_states = np.clip(next_states, low, high)

    return next_states, uncertainity


def mppi_action(x0, u_init=None, use_var=None):

    x0 = np.asarray(x0, dtype=np.float32)

    if u_init is None:
        u_init = np.zeros(H, dtype=np.float32)
    else:
        u_init = np.asarray(u_init, dtype=np.float32)
        assert u_init.shape[0] == H

    # Sample noise around current plan
    eps = rng.normal(0.0, SIGMA, size=(K, H)).astype(np.float32)
    U   = np.clip(u_init[None, :] + eps, act_low, act_high)

    # Rollout trajectories under GP model
    states = np.tile(x0, (K, 1))      # (K, 2)
    costs  = np.zeros(K, dtype=np.float32)

    for t in range(H):
        u_t = U[:, t]
        # compute uncertainty + *current* next state
        next_states, uncertainty = gp_step_batch(states, u_t)
        # cost of the transition from states -> next_states using u_t
        costs += stage_cost(next_states, u_t, uncertainty, use_var)
        states = next_states

    # MPPI weighting
    J_min   = costs.min()
    weights = np.exp(-(costs - J_min) / LAMBDA)
    weights_sum = weights.sum() + 1e-8

    # Weighted update of the control sequence
    du     = (weights[:, None] * eps).sum(axis=0) / weights_sum
    u_new  = np.clip(u_init + du, act_low, act_high)
    u0     = float(u_new[0])

    return u0, u_new
```

* A short discussion (2–3 paragraphs) explaining how including uncertainty affects performance and why it might lead to better results.

Adding uncertainty to the cost makes the controller more cautious. When the system penalize GP variance, is telling to the MPPI: “prefer trajectories where the model is confident, and avoid regions where the model hasn’t seen much data.” That’s good for robustness, because is less likely to follow a trajectory that looks great under a wrong model. But those “risky” regions might actually contain the fastest paths to the goal, so a very strong uncertainty penalty can make the agent more conservative, move less aggressively, and sometimes even fail to reach the goal.

So, including uncertainty is mainly about trading off performance vs model risk, not directly about speed or sample count. With limited training data, an uncertainty-aware controller can sometimes perform better overall than a mean-only controller, because it avoids areas where the model is unreliable. But there’s no guarantee it will reach the goal in fewer steps or with fewer sampled trajectories if the fastest routes pass through high-variance regions that the cost is now heavily penalizing.
